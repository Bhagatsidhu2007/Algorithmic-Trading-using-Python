```python
ret = aapl.pct_change().dropna()  # Calculate daily % returns and remove first NaN row
ret.info() # To check if null value was dropped
```

```python
ret.plot(kind="hist", figsize=(12,8), bins=100)  # Plot histogram of daily returns with 100 bins -> More bins more details
plt.show()  # Display the plot
#AAPL’s daily returns cluster near zero, showing it’s a stable, low-volatility stock.
```

```python
# Calculate mean daily return
daily_mean_ret = ret.mean()
print("Mean daily return:", daily_mean_ret)

# Calculate daily variance
var_daily = ret.var()
print("Daily variance:", var_daily)

# Calculate daily standard deviation (volatility)
std_daily = np.sqrt(var_daily)
print("Daily standard deviation:", std_daily)

# Alternative way to get daily std
print("Daily standard using ret.std():", ret.std())

# Annualize mean return (252 trading days)
annual_mean_ret = daily_mean_ret * 252 # 252 is the average number of trading days in a year
print("Annual mean return:", annual_mean_ret)

# Annualize variance
annual_var_return = var_daily * 252 # 252 is the average number of trading days in a year
print("Annual variance:", annual_var_return)

# Annual standard deviation (volatility)
annual_std_retuens = np.sqrt(annual_var_return)
print("Annual standard deviation:", annual_std_retuens)

# Another way to annualize std (using daily std)
print("Annual std using ret.std()*sqrt(252):", ret.std() * np.sqrt(252)) # 252 is the average number of trading days in a year
```

```python
# Return is how much money you make from an investment. Here, it's the average % gain per year for a stock.
# Risk here means how much the stock's return goes up and down which is also called volatility. This can be measured by standard deviation (std)

summary.plot.scatter(x="std", y="mean", figsize=(12,8), s=50, fontsize=15)  # Scatter plot: risk (std) vs return

for i in summary.index:
    # Add stock name labels near each point for clarity
    plt.annotate(i, xy=(summary.loc[i, "std"] + 0.002, summary.loc[i, "mean"] + 0.002), size=15)

plt.xlabel("Annual risk (std)", fontsize=15)  # Label x-axis
plt.ylabel("Annual return", fontsize=15)  # Label y-axis
plt.title("Risk/Return", fontsize=25)  # Add chart title
plt.show()  # Display the plot

"""Insights from the graph:
AAPL (Apple): Highest return and highest risk.
MSFT (Microsoft): High return with relatively lower risk than AAPL — strong risk-return profile.
SPY: Lower risk and moderate return.
KO (Coca-Cola): Low risk, low return.
IBM and DIS: Moderate to high risk but relatively low returns."""
```

### Daily Returns

**Definition:**  
Daily return shows the percentage change in price from one day to the next.  

**Why important:**  
It tells you how much the stock’s value changed on each day, which is the building block to calculate overall gains or losses.

---

### Cumulative Returns

**Definition:**  
Cumulative return shows total gain or loss over a period, accounting for compounding of daily returns.  

**Why important:**  
It shows how much your investment would have grown or shrunk if you bought at the start and held until now.

**Interpretation:**  
- A cumulative return of 0.5 means your investment grew by 50%.  
- A cumulative return of -0.2 means a 20% loss.


```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download Apple stock data
apple = yf.download("AAPL")  # DataFrame with OHLCV and more

# Extract 'Close' prices as Series
close_prices = apple["Close"]  

# Calculate daily simple returns (percentage change)
daily_returns = close_prices.pct_change().dropna()  # Drop first NaN

# Calculate cumulative returns for buy-and-hold strategy
cum_returns = (1 + daily_returns).cumprod() - 1   # Cumulative product of (1 + daily returns) minus 1

# Plot cumulative returns
plt.figure(figsize=(12, 6))                       # Figure size
plt.plot(cum_returns, label="Buy and Hold Cumulative Return")  # Plot cumulative returns
plt.title("Apple Buy and Hold Strategy - Cumulative Returns")  # Chart title
plt.xlabel("Date")                                # X-axis label
plt.ylabel("Cumulative Return")                   # Y-axis label
plt.grid(True)                                    # Grid for readability
plt.legend()                                      # Legend
plt.show()                                        # Display plot
```

```python
# Plot High, Low, and Cumulative Max of High
plt.figure(figsize=(15, 8))

# Plot High prices
plt.plot(apple["High"], label="High", color="blue")

# Plot Low prices
plt.plot(apple["Low"], label="Low", color="red")

# Plot cumulative maximum of High prices
plt.plot(apple["High"].cummax(), label="Cumulative Max (High)", color="green", linestyle="--")

# Add title and labels
plt.title("Apple Stock: High, Low, and Cumulative Max of High", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
"""
High: Daily high prices of Apple stock.

Low: Daily low prices of Apple stock.

Cumulative Max: The highest price seen so far at any point in time """
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download Apple stock data
data = yf.download("AAPL", start="2015-01-01", end="2025-01-01")

# Flatten column names if MultiIndex (as happens with yfinance sometimes)
data.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in data.columns]

# Calculate moving averages
data["SMA50"] = data["Close_AAPL"].rolling(window=50).mean()
data["SMA200"] = data["Close_AAPL"].rolling(window=200).mean()

# Create signal column
data["Signal"] = 0
data.loc[data["SMA50"] > data["SMA200"], "Signal"] = 1
data.loc[data["SMA50"] < data["SMA200"], "Signal"] = -1

# Plotting
plt.figure(figsize=(15, 8))
plt.plot(data["Close_AAPL"], label="Close Price", alpha=0.5)
plt.plot(data["SMA50"], label="SMA50", color="blue")
plt.plot(data["SMA200"], label="SMA200", color="red")

# Fill long and short areas
plt.fill_between(data.index, data["Close_AAPL"], where=(data["Signal"] == 1), color='green', alpha=0.1, label="Long Signal")
plt.fill_between(data.index, data["Close_AAPL"], where=(data["Signal"] == -1), color='red', alpha=0.1, label="Short Signal")

plt.title("AAPL: SMA50 vs SMA200 Strategy", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
""" SMA50 > SMA200 → Recent prices are higher than the long-term average → Uptrend → Go long (buy).

SMA50 < SMA200 → Recent prices are lower than the long-term average → Downtrend → Go short (sell). """
```

```python
# 1. Original SMA Crossover signal: 1 for long, -1 for short, 0 for neutral
data["Signal"] = 0
data.loc[data["SMA50"] > data["SMA200"], "Signal"] = 1
data.loc[data["SMA50"] < data["SMA200"], "Signal"] = -1

# Shift signal by 1 to avoid lookahead bias (act on next day)
data["Position"] = data["Signal"].shift(1)

# 2. Long Bias Strategy signal: 1 for long, 0 for cash (no shorting)
data["LongBias_Signal"] = 0
data.loc[data["SMA50"] > data["SMA200"], "LongBias_Signal"] = 1
data["LongBias_Position"] = data["LongBias_Signal"].shift(1)

# Calculate daily returns from closing price
data["Daily Return"] = data["Close_AAPL"].pct_change()

# Calculate strategy returns for SMA crossover strategy
data["Strategy Return"] = data["Position"] * data["Daily Return"]

# Calculate strategy returns for Long Bias strategy
data["LongBias_Return"] = data["LongBias_Position"] * data["Daily Return"]

# Calculate cumulative returns for buy & hold and both strategies
data["Cumulative Market Return"] = (1 + data["Daily Return"]).cumprod()
data["Cumulative Strategy Return"] = (1 + data["Strategy Return"]).cumprod()
data["Cumulative LongBias Return"] = (1 + data["LongBias_Return"]).cumprod()

# Plot all cumulative returns
plt.figure(figsize=(15, 8))
plt.plot(data["Cumulative Market Return"], label="Buy & Hold", color="black")
plt.plot(data["Cumulative Strategy Return"], label="SMA Crossover Strategy", color="orange")
plt.plot(data["Cumulative LongBias Return"], label="Long Bias Strategy", color="green")
plt.title("Cumulative Returns: Buy & Hold vs SMA Crossover vs Long Bias", fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
